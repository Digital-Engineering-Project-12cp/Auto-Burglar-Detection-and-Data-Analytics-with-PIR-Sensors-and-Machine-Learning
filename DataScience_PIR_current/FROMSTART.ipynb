```python
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
```

```python
centraldf['DATE_TIME'] = pd.to_datetime(centraldf['DATE_TIME'], dayfirst=True)
```

```python
centraldf.to_csv('reference.csv')
```

```python
centraldf["DAY_TYPE"] = centraldf.DATE_TIME.apply(lambda x: 1 if x.dayofweek > 5 else 0  )
```

```python
#FINDING FEDERAL HOLIDAYS
```

```python
cal = calendar()
holidays = cal.holidays(start = centraldf.DATE_TIME.min(), end = centraldf.DATE_TIME.max())
centraldf["IS_HOLIDAY"] = centraldf.DATE_TIME.isin(holidays)
```

```python
for obs in range(1,6):
    centraldf["T_" + str(obs)] = centraldf.Motion.shift(obs)
```

```python
centraldf.fillna(0.00,inplace=True)
centraldf.head(10)
```

```python
centraldf.IS_HOLIDAY = centraldf.IS_HOLIDAY.astype("int")
```

```python
clean_data = centraldf[['DAY_TYPE', 'IS_HOLIDAY', 'T_1','T_2', 'T_3', 'T_4', 'T_5','Motion']]
clean_data.head(10)
```

# Setting Training and Test Data

```python
training_data = centraldf[centraldf.DATE_TIME < pd.to_datetime("06/03/2019")]
```

```python
training_data.tail(5)
```

```python
val_mask = (centraldf.DATE_TIME >= pd.to_datetime("06/03/2019")) & (centraldf.DATE_TIME < pd.to_datetime("06/05/2019"))
val_data = centraldf.loc[val_mask]
```

```python
val_data.tail(5)
```

```python
test_data = centraldf[centraldf.DATE_TIME >= pd.to_datetime("06/05/2019")]
```

```python
test_data.tail(5)
```

```python
clean_train = training_data[['DAY_TYPE', 'IS_HOLIDAY', 'T_1','T_2', 'T_3', 'T_4', 'T_5','Motion']]
clean_test = test_data[['DAY_TYPE', 'IS_HOLIDAY', 'T_1','T_2', 'T_3', 'T_4', 'T_5','Motion']]
clean_val = val_data[['DAY_TYPE', 'IS_HOLIDAY', 'T_1','T_2', 'T_3', 'T_4', 'T_5','Motion']]
```

```python
clean_train.count()
```

```python
clean_test.count()
```

# LETS MODEL LSTM AND PREDICT

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train,y_train = clean_train.drop(["Motion"],axis=1),clean_train.Motion
X_test,y_test = clean_test.drop(["Motion"],axis=1),clean_test.Motion
X_val,y_val = clean_val.drop(["Motion"],axis=1),clean_val.Motion
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
X_valid_scaled = scaler.fit_transform(X_val)
X_train_scaled.shape[0]
```

```python
X_train_scaled.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model_k = Sequential()
model_k.add(LSTM(1, input_shape=(1,7)))
model_k.add(Dense(1))
model_k.compile(loss='mean_squared_error', optimizer='adam')
```

```python
X_t_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
```

```python
X_val_resaped = X_valid_scaled.reshape((X_valid_scaled.shape[0], 1, X_valid_scaled.shape[1]))
```

```python
history = model_k.fit(X_t_reshaped, y_train, validation_data=(X_val_resaped, y_val),\
epochs=10, batch_size=96, verbose=2)
```

```python
X_te_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))
```

```python
res = model_k.predict(X_te_reshaped)
```

```python
test_data["DL_PRED"] = res
```

```python
test_data.plot(x="DATE_TIME",y=["Motion","DL_PRED"])
plt.show()
```
